```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
x=np.array([0,1,2,3])
y=np.array([5,10,33,40])
plt.subplot(1,2,2)    #row,col,pos
plt.plot(x,y)
plt.subplot(1,2,1)
plt.scatter(x,y)
plt.show()
```

```python
x=np.array([0,1,2,3])
y=np.array([5,10,33,40])
plt.subplot(2,1,1)    #row,col,pos
plt.plot(x,y)
plt.subplot(2,1,2)
plt.scatter(x,y)
plt.show()
```

```python
x=np.array([0,1,2,3])
y=np.array([5,10,33,40])
plt.subplot(2,2,1)    #row,col,pos
plt.plot(x,y)
plt.subplot(2,2,2)
plt.scatter(x,y)
plt.subplot(2,2,3)
plt.pie(x)
plt.subplot(2,2,4)
plt.hist(x)
plt.show()
```

```python
# *SEABORN*
import seaborn as sns
tips=sns.load_dataset('tips')
print(tips)
sns.countplot(x='sex',data=tips)
plt.show()
```

```python
sns.countplot(x='smoker',data=tips)
```

```python
plt.figure(figsize=(12,4))      #to change size using matplotlib
sns.countplot(x='time',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,aspect=1)#aspect is used in seaborn to change size ie zoom in or zoom out
```

```python
sns.set_context('poster') #for big size
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context('paper')
sns.countplot(x='sex',data=tips)
```
