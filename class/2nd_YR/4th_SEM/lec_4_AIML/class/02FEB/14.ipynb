```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
sns.set_style('ticks')#white,dark , whitegrid, darkgrid, ticks
sns.countplot(x='sex',data=tips)
```

```python
fmri=sns.load_dataset('fmri')
print(fmri)
```

```python
sns.scatterplot(x='timepoint',y='signal',hue='region',style='event',data=fmri)#
```

```python
tips=sns.load_dataset('tips')
sns.histplot(tips['total_bill'])
plt.show()
```

```python
tips=sns.load_dataset('tips')
sns.histplot(tips['total_bill'],kde=True)#kernel density estimator, show density of increase in graph,cannot go up than highest plot part of graph
plt.show()
```

```python
#visualise the dataset penguins on bodymass 
peg=sns.load_dataset('penguins')
print(peg.to_string())
sns.scatterplot(peg['body_mass_g'])
plt.show()
```

```python
#visualise the dataset penguins on bodymass 
peg=sns.load_dataset('penguins')
print(peg.to_string())
sns.histplot(peg['body_mass_g'])
plt.show()
```

```python
peg=sns.load_dataset('penguins')
sns.histplot(data=peg,x='body_mass_g',kde=True,hue='species')# divides data into count of colors with diff element in species
plt.show()
```

```python
sns.countplot(data=peg,x='species')
plt.show()
```
