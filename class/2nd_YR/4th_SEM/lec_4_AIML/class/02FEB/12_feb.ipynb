```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array(['a','b','c','d'])
y=np.array([3,10,5,12])
plt.bar(x,y)
plt.show()
```

```python
x=np.array(['a','b','c','d'])
y=np.array([3,10,5,12])
plt.barh(x,y)
plt.show()
```

```python
x=np.array(['a','b','c','d'])
y=np.array([3,10,5,12])
plt.barh(x,y,height=0.3)
plt.show()
```

```python
x=np.array(['a','b','c','d'])
y=np.array([3,10,5,12])
plt.bar(x,y,width=0.3)
plt.show()
```

```python
x=np.random.normal(200,20,300)
plt.hist(x)
plt.show()
```
