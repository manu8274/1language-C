```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=np.array([3,8,1,10])
b=np.array([6,2,7,11])
plt.scatter(a,b)
plt.show()
```

```python
a=np.array([3,8,1,10])
b=np.array([6,2,7,11])
color=np.array(['red','green','orange','brown'])
plt.scatter(a,b,c=color)
plt.show()
```

```python
a=np.array([3,8,1,10])
b=np.array([6,2,7,11])
color=np.array(['red','green','orange','brown'])
size=np.array([300,2000,100,800])
plt.scatter(a,b,c=color,s=size)
plt.show()
```

```python
a=np.array([3,8,1,10])
b=np.array([6,2,7,11])
color=np.array([1,2,3,4])
size=np.array([300,2000,100,800])
plt.scatter(a,b,c=color,s=size)
plt.show()
```

```python
a=np.array([3,8,1,10])
b=np.array([6,2,7,11])
color=np.array([1,2,3,4])
plt.scatter(a,b,c=color,cmap='spring')
plt.colorbar()
plt.show()
```

```python
a=np.array([3,8,1,10])
b=np.array([6,2,7,11])
color=np.array([1,2,3,4])
plt.scatter(a,b,c=color,cmap='Greens')
plt.colorbar()
plt.show()
```

```python
import random
a=np.random.randint(100,size=(100))
b=np.random.randint(100,size=(100))
color=np.random.randint(100,size=(100))
size=np.random.randint(100,size=(100))*10
plt.scatter(a,b,c=color,s=size,cmap='nipy_spectral')
plt.colorbar()
plt.show()
```
