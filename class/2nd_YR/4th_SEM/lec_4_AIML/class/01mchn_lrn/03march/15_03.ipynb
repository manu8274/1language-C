```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
data=pd.read_csv("homeprices.csv")
print(data)
```

```python
plt.scatter(data["area"],data["price"])
plt.xlabel("area")
plt.ylabel("prices")
plt.show()
```

```python
ar=data.drop("price",axis="columns")
pr=data.drop("area",axis="columns")
print(ar,pr)
```

```python
from sklearn.model_selection import train_test_split
# x1,y1=train_test_split(ar,pr)# no to split as dataset is small
x1,y1=ar,pr
print(x1)
print(y1)
```

```python
from sklearn import linear_model
lr=linear_model.LinearRegression()
lr.fit(x1,y1)
```

```python
plt.scatter(data["area"],data["price"])
plt.plot(x1,lr.predict(x1),color="black")
# plt.plot(y1,lr.predict(y1),color="red")
plt.show()
```

```python
x=(lr.score(x1,y1))
print("accuracy = ",round((x*100),3),"%")
```

```python
# lr.predict([[3300,122]])
print(lr.predict([[3300]]))
print(lr.predict([[5000]]))
```
