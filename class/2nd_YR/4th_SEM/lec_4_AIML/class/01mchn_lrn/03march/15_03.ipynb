```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
data=pd.read_csv("homeprices.csv")
print(data)
```

```python
plt.scatter(data["area"],data["price"])
plt.xlabel("area")
plt.ylabel("prices")
plt.show()
```

```python
ar=data.drop("price",axis="columns")
pr=data.drop("area",axis="columns")
print(ar,pr)
```

```python
from sklearn.model_selection import train_test_split
# x1,y1=train_test_split(ar,pr)# no to split as dataset is small
x1,y1=ar,pr
print(x1)
print(y1)
```

```python
from sklearn import linear_model
lr=linear_model.LinearRegression()
lr.fit(x1,y1)
```

```python
plt.scatter(data["area"],data["price"])
plt.plot(x1,lr.predict(x1),color="black")
# plt.plot(y1,lr.predict(y1),color="red")
plt.show()
```

```python
x=(lr.score(x1,y1))
print("accuracy = ",round((x*100),3),"%")
```

```python
# lr.predict([[3300,122]])
print(lr.predict([[3300]]))
print(lr.predict([[5000]]))
```

```python
df=pd.read_csv("canada_per_capita_income.csv")
print(df)
```

```python
plt.scatter(df["year"],df["per capita income (US$)"])
plt.show()
```

```python
pc=df["per capita income (US$)"]
yr=df.drop("per capita income (US$)" ,axis="columns")
print(pc)
print(yr)
```

```python
# from sklearn.model_selection import train_test_split
x1,y1=yr,pc
print(x1,y1)
```

```python
from sklearn import linear_model
lrn=linear_model.LinearRegression()
lrn.fit(x1,y1)
```

```python
print(round(lrn.score(x1,y1)*100,5),"%")
```

```python
print("2020",lrn.predict([[2020]]))
print("2021",lrn.predict([[2021]]))
print("2015",lrn.predict([[2015]]))
print("2000",lrn.predict([[2000]]))
print("1998",lrn.predict([[1998]]))
print("1974",lrn.predict([[1974]]))
```

```python
plt.scatter(df["year"],df["per capita income (US$)"])
plt.plot(x1,lrn.predict(x1),color="red")
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.xticks(rotation=90)
plt.show()
```

```python
m=lrn.coef_
c=lrn.intercept_
print("slope =",m,"intercept = ",c)
```
