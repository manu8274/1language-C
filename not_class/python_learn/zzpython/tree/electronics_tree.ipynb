```python
class treenode:
    def __init__(self,data):
        self.data=data
        self.children=[]
        self.parent=None
    
    def addchild(self,child):#to add child
        child.parent=self
        self.children.append(child)
    
    def printtree(self):#simple print
        print(self.data)
        for c in self.children:
            print("\t",c.data)
            for i in c.children:
                print("\t\t",i.data)
    
    def get_level(self):# to get level og root or leaves
        level=0
        p=self.parent
        while p:
            level+=1
            p=p.parent
        return level

    def print_tree_recursion(self): # to print through recursion
        print("\t"*(self.get_level()) + self.data) # to get spaces for acc to level
        if(self.children):
            for c in self.children:
                c.print_tree_recursion()

    def print_tree_prefix_recursion(self): # to print through recursion
        spaces=' ' * self.get_level()*1
        prefix = spaces
        if self.parent:
            prefix+="|__" 
        else:
            prefix+=""
        print(prefix+self.data) # to get spaces for acc to level
        if(self.children):
            for c in self.children:
                c.print_tree_prefix_recursion()

def buildtree():# to build tree
    root=treenode("electronics")
    
    laptop=treenode("laptop")
    laptop.addchild(treenode("MAC"))
    laptop.addchild(treenode("THINKPAD"))
    laptop.addchild(treenode("ASUS"))
    
    mobile=treenode("mobile")
    mobile.addchild(treenode("SAMSUNG"))
    mobile.addchild(treenode("MI"))
    mobile.addchild(treenode("IPHONE"))

    tv=treenode("TV")
    tv.addchild(treenode("SAMSUNG"))
    tv.addchild(treenode("PANASONIC"))

    root.addchild(laptop)
    root.addchild(mobile)
    root.addchild(tv)

    return root

if __name__=="__main__":
    root=buildtree()
    # root.printtree()
    # root.print_tree_recursion()
    root.print_tree_prefix_recursion()
```
