```python
class Node():# to create or add data 
    def __init__(self,data=None,next=None):
        self.data=data
        self.next=next
class linkedlist(): ## to create a linked list and operations on it
    def __init__(self):#to create a parameter self for class linkedlist()
        self.head=None
    def insert_at_begin(self,data):#to insert at begin 
        node=Node(data,self.head)
        self.head=node
        return
    def print(self): #to print all linked list
        if(self.head==None):
            print("linked list is empty")
            return
        itr=self.head
        llstr=''
        while(itr):
            llstr = llstr + str(itr.data) + ("-->")
            itr=itr.next
        print(llstr)
    def insert_at_end(self,data):#to insert at end of ll
        if(self.head is None):
            self.head=Node(data,None)
            return
        itr=self.head
        while(itr.next):
            itr=itr.next
        itr.next=Node(data , None)
    def get_length(self): # to get length of the ll
        count=1
        itr=self.head
        while(itr.next):
            count+=1
            itr=itr.next
        return count

    def insert_at_index(self,index,data): #to insert at given index
        if(index<0 or index>=self.get_length()):
            raise Exception("error")
        if(index==0):
            self.insert_at_begin(data)
        itr=self.head
        count=0
        while(itr.next):
            if(count==index-1):
                itr.next=Node(data,itr.next)
                break
            count+=1
            itr=itr.next
    def remove_at(self,index): # to remove a cretain index
        if(index<0 or index>=self.get_length()):
            raise Exception("error")
        if(index==0):
            self.head=self.head.next
            return 
        itr=self.head
        c=0
        while itr:
            if(index-1==c):
                itr.next=itr.next.next
            itr=itr.next
            c+=1
    def insert_val_from_lst(self,data): # to insert values from list
        self.head=None
        for data1 in data:
            self.insert_at_end(data1)

    def search_item(self,data):# to return  index of searched data
        itr=self.head
        c=0
        while(itr):
            if(itr.data==data):
                return c
            itr=itr.next
            c+=1
    # def search_str_item(self,data):# to return  index of searched string data
    #     itr=self.head
    #     c=0
    #     while(itr):
    #         if(itr.data==data):
    #             return c
    #         itr=itr.next
    #         c+=1

    def remove_by_value(self,data):
        itr=self.head
        while(itr):
            if(itr.next.data==data):
                itr.next=itr.next.next
                break
            itr=itr.next
if __name__=='__main__':
    ll=linkedlist()
    ll.insert_at_begin(5)
    ll.insert_at_begin(50)
    ll.insert_at_end(100)
    ll.print()
    ll.insert_at_index(2,109976)
    ll.print()
    ll.remove_at(1)
    print("after removing index 1 :",end="")
    ll.print()
    print("length of linked list:",ll.get_length())
    ll.insert_at_index(2,"manu")
    # ll.insert_val_from_lst([1,2,3,4])
    print("after inserting element \"manu\" at 2 index:",end=" ")
    ll.print()
    print("index of 100 is :",ll.search_item(100))
    print("index of manu is :",ll.search_item("manu"))
    ll.remove_by_value("manu")
    print("after removing by value \"manu\" : ",end=" ")
    ll.print()
```
